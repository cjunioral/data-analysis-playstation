```python
df.drop(['Game', 'Developer', 'Total Shipped', 'NA Sales', 'PAL Sales', 'Japan Sales',
          'Other Sales', 'Last Update', 'rating', 'ratings_count', 
          'metacritic', 'platforms'], axis=1, inplace=True)
df.head()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
tabela_tratada = df.to_csv('tabela_tratada.csv', index=False)
tabela_tratada
```

```python
vendas_por_console = df.groupby('Console')['Total Sales'].sum().to_frame()
vendas_por_console
```

```python
ordem_consoles = df.groupby('Console')['Total Sales'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(7, 3))
plt.barh(ordem_consoles.index, ordem_consoles.values) 

plt.title('Vendas por Console')
plt.xlabel('Total de Vendas (Milhes)')
plt.ylabel('Console')

plt.show()
```

```python
df.head()
```

```python
analise_publisher = df.groupby('Publisher').agg(Total_Vendas = ('Total Sales', 'sum'),
                                                Num_Jogos = ('Name', 'count')
                                                ).sort_values(by='Total_Vendas', ascending=False)

analise_publisher.head()
```

```python
plt.figure(figsize=(10, 6))
analise_publisher.head().plot(
    kind='bar',
    title='Top 5 Editoras por Vendas Totais',
    xlabel='Editora',
    ylabel='Total de Vendas (em Milhes)'
)
plt.xticks(rotation=45)
plt.show()
```
